# **Image to Pencil Sketch with Python**

### *Steps Involved:*

1. **Read the image in RBG format and then convert it to a grayscale image:** This will turn an image into a classic black and white photo. 
2. **Invert the grayscale image also called negative image, this will be our inverted grayscale image:** Inversion can be used to enhance details. 
3. **Create the pencil sketch by mixing the grayscale image with the inverted blurry image:** This can be done by dividing the grayscale image by the inverted blurry image.

*Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.*

```python
from PIL import Image
from IPython.display import display
import cv2  
```

```python
# Creating heading color-Purple, style- Bold, aling- Right shift
print('\033[1;35m PHOTO TO PENCIL SKETCH\n\033[0m'.center(20))

# Creating Options color- Canon, style -bold, normal
print('\033[1;36m Enter the path of the image: \033[0m')
img= input()
```

```python
step1 = cv2.imread(img ,1)                          # Loding the image to variable step1 
step1 =cv2.resize(step1 , (200,200))                # Resizing the image
step1 = cv2.cvtColor(step1 , cv2.COLOR_BGR2RGB)     # Setting up normal color for the image to be displayed

# Displaying Image--  
print('\033[1;35m\n Entered Image\n \033[0m'.center(10))
display(Image.fromarray(step1 ))
```

```python
#Converting the image to gray scale
step2 = cv2.cvtColor(step1 , cv2.COLOR_BGR2GRAY)

# Displaying Gray Image--  
print('\033[1;35m\n Gray Image\n \033[0m'.center(10))
display(Image.fromarray(step2 ))
```

```python
# Converting the gray image to inverted image
step3 = cv2.bitwise_not(step2 )

# Displaying Inverted Image--  
print('\033[1;35m\n Inverted Image\n \033[0m'.center(10))
display(Image.fromarray(step3 ))
```

```python
# Smoothening the inverted image
step4 = cv2.GaussianBlur(step3 , (5,5),0,0)

# Displaying Smooth Inverted Image--  
print('\033[1;35m\n Smooth Inverted Image\n \033[0m'.center(10))
display(Image.fromarray(step4 ))
```

```python
# Converting the smooth inverted image to Pencil sketch
step5 = 255- step4
final= cv2.divide(step2 , step5, scale=256.0)

# Displaying Pencil Sketch--  
print('\033[1;35m\n Pencil Sketch\n \033[0m'.center(10))
display(Image.fromarray(final))
```

```python
# Displaying entered image and the final outcome together

print('\033[1;35m\n Entered Image: \033[0m'.center(10))
display(Image.fromarray(step1))
print('\033[1;35m\n Final Sketch:\033[0m'.center(10))
display(Image.fromarray(final))
```
